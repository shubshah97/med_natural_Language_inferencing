```python
def classCounts(df):
    grp=df.groupby(['gold_label'])['id'].nunique()
    return {key:grp[key] for key in list(grp.keys())}
```

```python
def classProportions(df):
    classC=classCounts(df)
    return {val[0]: round(val[1]/df.shape[0],4) for val in classC.items()}
```
