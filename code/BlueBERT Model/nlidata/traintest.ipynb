```python
dev_files = [os.path.join(os.getcwd(),"nliraw/dev",i) for i in os.listdir(os.path.join(os.getcwd(),"nliraw/dev"))]
test_files = [os.path.join(os.getcwd(),"nliraw/test",i) for i in os.listdir(os.path.join(os.getcwd(),"nliraw/test"))]
```

```python
columnsneeded = ['gold_label','sentence1', "sentence2"]
train_dataframe= pd.DataFrame([], columns = columnsneeded)
for i in train_files:
    temp_dataFrame = pd.read_json(i,lines = True)
    train_dataframe = pd.concat([train_dataframe,temp_dataFrame])
train_dataframe=train_dataframe.reset_index()
```

```python
columnsneeded = ['gold_label','sentence1', "sentence2"]
test_dataframe= pd.DataFrame([], columns = columnsneeded)
for i in test_files[:-1]:
    temp_dataFrame = pd.read_json(i,lines = True)
    test_dataframe = pd.concat([test_dataframe,temp_dataFrame])
test_dataframe = test_dataframe.reset_index()
```

```python
columnsneeded = ['gold_label','sentence1', "sentence2"]
dev_dataframe= pd.DataFrame([], columns = columnsneeded)
for i in dev_files:
    temp_dataFrame = pd.read_json(i,lines = True)
    dev_dataframe = pd.concat([dev_dataframe,temp_dataFrame])
dev_dataframe = dev_dataframe.reset_index()
```

```python
train_dataframe = train_dataframe[train_dataframe.gold_label.isin(['entailment', 'contradiction', "neutral"])]
test_dataframe = test_dataframe[test_dataframe.gold_label.isin(['entailment', 'contradiction', "neutral"])]
dev_dataframe = dev_dataframe[dev_dataframe.gold_label.isin(['entailment', 'contradiction', "neutral"])]
```
